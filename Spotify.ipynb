```python
import pandas as pd

df = pd.read_csv("spotify.csv")

df.head(5)
```

```python
# procura por valores nulos
df.isnull().sum()
```

```python
# remove as linhas com valores nulos
df = df.dropna()
df.isnull().sum()
```

```python
df.dtypes
```

```python
#escolhe popularity como alvo de nossa análise
dfX = df.drop(['popularity', 'track_id', 'artists', 'album_name', 'track_name', 'Unnamed: 0'], axis=1)
dfY= df['popularity']

dfX.shape
```

```python
dfX['explicit'].value_counts().plot(kind='bar')
```

```python
import matplotlib.pyplot as plt

plt.plot(dfX['duration_ms'], dfY)
plt.show()
```

```python
plt.plot(dfX['danceability'], dfY)
plt.show()
```

```python
plt.plot(dfX['energy'], dfY)
plt.show()
```

```python
plt.plot(dfX['loudness'], dfY)
plt.show()
```

```python
plt.plot(dfX['speechiness'], dfY)
plt.show()
```

```python
# transforma dados categóricos em numéricos
dfX = pd.get_dummies(dfX)
dfX.head(5)
```
